```python
import guppylang
from guppylang import guppy

from typing import Counter

from guppylang import guppy
from guppylang.std.builtins import result, array, exit, panic
from guppylang.std.quantum import measure_array, measure, qubit, h, cx
from guppylang.defs import GuppyFunctionDefinition

from selene_sim import DepolarizingErrorModel
from selene_sim.exceptions import SelenePanicError

from guppylang.defs import GuppyFunctionDefinition
from guppylang.std.quantum import qubit, measure, h, cx


def run(main_def: GuppyFunctionDefinition) -> Counter:
    res = (
        main_def.emulator(n_qubits=8)
        .stabilizer_sim()
        .with_seed(42)
        .with_shots(1)
        .run()
    )

    return res.collated_counts()
```

```python
from guppylang.std.builtins import result

@guppy.struct
class Name:
    text: str

@guppy.struct
class QubitPair:
    q1: int
    q2: int
    name: str

    @guppy
    def __init__(self: "QubitPair", q1: int, q2: int) -> None:
        self.q1 = q1
        self.q2 = q2*2
        self.name = "pair"

    @guppy
    def method(self: "QubitPair") -> bool:
        return self.q1 == self.q2


print("§§§§§§\n\n")
# QubitPair.check()



@guppy
def make_struct() -> None:
    # a,b = qubit(), qubit()
    pair = QubitPair(1, 2, "prova")
    
    
    # pair.method()
    result("tag", pair.method())



# run(make_struct)
print(make_struct.compile_function())
```

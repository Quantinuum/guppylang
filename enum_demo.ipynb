```python
import guppylang
from guppylang import guppy


from guppylang.std.quantum import qubit, measure, h, cx
```

```python
@guppy.struct
class Name:
    text: str

@guppy.struct
class QubitPair:
    q1: qubit
    q2: qubit
    name: QubitPair

    @guppy
    def method(self: "QubitPair") -> None:
        h(self.q1)
        cx(self.q1, self.q2)
```

```python

from guppylang.decorator import guppy
from guppylang.std.qsystem import measure
from guppylang.std.quantum import h, qubit, x, rz, angle
from guppylang.std.builtins import owned
from guppylang.std.modifier import dagger, control


import guppylang
guppylang.enable_experimental_features()

# Dummy variables to suppress Undefined name
# TODO: `ruff` fails when without these, which need to be fixed

daggera = object()

@guppy
def discard(a: qubit @ owned) -> None:
    measure(a)

@guppy
def bar(q1: qubit, q2: qubit) -> None:
        rz(q1, angle(0.1))
        with daggera():
             rz(q1, angle(0.1))
        with control(q1):
            x(q2)

bar.compile_function().modules[0].render_dot()

```

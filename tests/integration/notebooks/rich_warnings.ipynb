```python
from dataclasses import dataclass
from typing import ClassVar

from guppylang import rich_warnings
from guppylang_internals.diagnostic import Warning
from guppylang_internals.engine import DEF_STORE
from guppylang_internals.error import diagnostic_report, emit_warning
from guppylang_internals.span import Loc, Span
```

```python
import warnings

@dataclass(frozen=True)
class NotebookWarning(Warning):
    title: ClassVar[str] = "Notebook warning"
    span_label: ClassVar[str] = "This warning was rendered from a notebook"


file = "rich_warning_notebook.guppy"
DEF_STORE.sources.add_file(file, "suspicious_code()\n")

with warnings.catch_warnings(record=True) as records:
    warnings.simplefilter("always")
    with rich_warnings(), diagnostic_report():
        emit_warning(NotebookWarning(Span(Loc(file, 1, 0), Loc(file, 1, 17))))

assert len(records) == 1
```
